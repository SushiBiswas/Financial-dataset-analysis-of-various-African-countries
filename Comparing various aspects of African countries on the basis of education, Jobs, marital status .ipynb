```python
age_column = df['Respondent Age']
```

```python
total_values = age_column.count()
```

```python
# Median
median_age = age_column.median()

# Mean
mean_age = age_column.mean()

# Minimum and Maximum values
min_age = age_column.min()
max_age = age_column.max()

# Standard variance
std_variance = age_column.std()

# Quartiles (25%, 50%, 75%)
quartiles = age_column.quantile([0.25, 0.5, 0.75])

print("Count of total values:", total_values)
print("Median:", median_age)
print("Mean:", mean_age)
print("Minimum value:", min_age)
print("Maximum value:", max_age)
print("Standard variance:", std_variance)
print("Quartiles (25%, 50%, 75%):")
print(quartiles)
```
